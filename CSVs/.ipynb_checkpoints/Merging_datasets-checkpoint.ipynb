```python
import pandas as pd
```

```python
df = pd.read_csv('./StormEvents_details_2022.csv')
df.head()
```

```python
df.info()
```

```python
df2 = pd.read_csv('./StormEvents_fatalities_2022.csv')
df2.head()
```

```python
df.info()
```

```python
df1 = pd.read_csv('./StormEvents_locations_2022.csv')
df1.head()
```

```python
df.info()
```

```python
first_merge = pd.merge(df, df1, on="EVENT_ID", how='left')
first_merge.head()
```

```python
second_merge = pd.merge(first_merge, df2, on='EVENT_ID', how='left')
second_merge.head()
```

```python
df_2023d = pd.read_csv('./StormEvents_details_2023.csv')
df_2023d.head()
```

```python
df_2023d.info()
```

```python
df_2023f = pd.read_csv('./StormEvents_fatalities_2023.csv')
df_2023f.info()
```

```python
df_2023l = pd.read_csv('./StormEvents_locations_2023.csv')
df_2023l.info()
```

```python
first_join = pd.merge(df_2023d, df_2023f, on="EVENT_ID", how='left')
first_join.head()
```

```python
second_join = pd.merge(first_join, df_2023l, on='EVENT_ID', how='left')
second_join.head()
```

```python
final_df = pd.concat([second_merge, second_join])
final_df.info()
```

```python
final_df.to_csv("stormdata.csv", index=False)
```

```python
pd.set_option('display.max_columns', None)
```

```python
column_drop_df = final_df.drop(columns=['WFO','BEGIN_DAY', 'BEGIN_TIME', 'END_YEARMONTH',\
                                        'END_DAY','END_TIME','BEGIN_DATE_TIME','END_DATE_TIME',\
                                        'CZ_TYPE', 'CZ_FIPS', 'STATE_FIPS', 'CZ_TIMEZONE', 'MAGNITUDE',\
                                        'MAGNITUDE_TYPE','SOURCE', 'CATEGORY', 'TOR_OTHER_WFO', 'TOR_OTHER_CZ_STATE',\
                                        'TOR_OTHER_CZ_FIPS', 'TOR_OTHER_CZ_NAME', 'BEGIN_AZIMUTH', 'YEARMONTH',\
                                        'END_AZIMUTH', 'DATA_SOURCE', 'EPISODE_ID_y', 'LOCATION_INDEX', 'LOCATION',\
                                        'RANGE', 'AZIMUTH', 'LAT2', 'LON2', 'FAT_YEARMONTH', 'FAT_DAY', 'FAT_TIME',\
                                        'FATALITY_DATE', 'FATALITY_AGE','FATALITY_SEX', 'EVENT_YEARMONTH','FLOOD_CAUSE',\
                                       'TOR_F_SCALE','TOR_WIDTH','LATITUDE','LONGITUDE'])
```

```python
column_drop_df['MONTH'] = column_drop_df['BEGIN_YEARMONTH']-(column_drop_df['YEAR']*100)
column_drop_df.head()
```

```python
column_drop_df = column_drop_df.drop(columns=['BEGIN_YEARMONTH'])
column_drop_df.head()
```

```python
column_drop_df.to_csv("cleaned_stormdata.csv", index=False)
```
